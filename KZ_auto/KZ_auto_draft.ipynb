**Содержание столбцов**
Название столбца исчерпывающе указывает на смысл и содержание каждого столбца. Год и месяц указывают на дату продажи. Компания - то же что и наименование дилерского центра, совершившего продажу. К характеристикам авто (т.е. продаваемого продукта) относятся столбцы бренд, модель, модификация, год выпуска, страна-производитель (+локализация производства), вид топлива, объем двигателя, тип коробки передач, тип привода, а также сегмент и класс. Важность столбцов для дальнейшего анализа рассматривается ниже.

**Пропуски**

Всего пропуски наблюдаются в 10 столбцах, в большинстве случаев доля пропусков в столбце составляет порядка 10-15%. В основном пропуски присутствуют в столбцах, которые *связаны с другими столбцами* (это например, бренд авто и его марка; бренд+марка и такие характеристики как объем двигателя, коробка передач, тип привода) - в таком случае данные частично можно попробовать восстановить с учетом данных из других столбцов. Пропуски в количестве авто и годе выпуска не зависят от значений в других столбцах и восстановить их невозможно.

В столбце "форма расчета" почти 2/3 пропусков, что может говорить о том, что не все дилерские центры собирают такую информацию. Такие пропуски можно отнести к типу MAR. Пропуски в типе клиента тоже можно отнести к типу MAR предположив также, что некоторые дилерские центры также не собирают данную информацию или работают только с одним типом клиентов (только физ. лица, например). К пропускам типа MAR можно отнести пропуски в объеме двигателя для электрокаров.

Пропуски в других столбцах можно отнести к типу MCAR.

**Объединение года и месяца**

Для начала нужно название месяца перевести в номер. Чтобы делать это не вручную используем локаль.

```python
locale.setlocale(locale.LC_TIME, '')
```

```python
kz['месяц'].unique()
```

Получили номер месяца, теперь создадим новый столбец с датой, в которой соединим год, месяц и предварительно укажем первый день месяца.

```python
kz['дата'] = pd.to_datetime(kz['год'].astype(str) + '/' + kz['месяц'].astype(str) + '/01')
```

```python
def last_day(d):
    '''Замена дня в дате на последний день соответствующего месяца'''
    
    next_month = d.replace(day=28) + dt.timedelta(days=4) #меняем везде день на 28 (последний день в феврале), 
                                                          #а прибавив 4 дня в любом случае окажемся в следующем месяце
    return next_month - dt.timedelta(days=next_month.day)
```

```python
kz['дата'] = kz['дата'].apply(last_day)
```

**Кодирование вида топива**

```python
kz['вид_топлива'].unique()
```

Видим несколько некорректных значений (числа). Посмотрим на них детальнее.

```python
kz.query('вид_топлива == "2" or вид_топлива == "1,6" or вид_топлива == "0"')
```

Судя по всему здесь произошел "сдвиг" по столбцам при записи данных (для тех строк где вид топлива 2 или 1,6). Так в столбце с видом топлива оказался объем двигателя, а в объеме двигателя указан вид коробки передач. Исправим это. Для начала выпишем индексы таких ошибочных строк в отдельный список.

```python
for_shift = kz.query('вид_топлива == "2" or вид_топлива == "1,6"').index
```

```python
kz.loc[for_shift, ['вид_топлива', 'объем_двиг_л', 'коробка_передач']] = (
    kz.loc[for_shift, ['вид_топлива', 'объем_двиг_л', 'коробка_передач']].shift(periods=1, axis=1)
)
```

```python
kz.loc[for_shift]
```

Данные оказались в своих столбцах. Можно присутпать к категоризации вида топлива. Будем использовать следующие значений: F - fuel, D - diesel, E - electro, HYB - hybrid. Но сначала приведем к нижнему регистру и удалим ненужные пробелы.

```python
kz['вид_топлива'] = (kz['вид_топлива'].str.lower()
                     .str.strip()
                     .str.replace('бензин', 'F')
                     .str.replace('дизель', 'D')
                     .str.replace('гибрид', 'HYB')
                     .str.replace('электро', 'E')
                     .str.replace('электричество', 'E'))
```

```python
kz['вид_топлива'].unique()
```

Категоризацию выполнили, однако остались пропуски и значение 0 - возможно их можно заполнить модой с учетом бренда и модели авто. Проверим, если ли модель с более чем 1 видом топлива и если есть, то сколько их.

```python
fuel_df = (kz.groupby(['бренд', 'модель'], as_index=False)
           .agg({'вид_топлива': 'nunique'})
           .query('вид_топлива > 1')
           .reset_index(drop=True))
len(fuel_df)
```

Выполним замену на наиболее частое значение по столбцу с видом топлива.

```python
for i in fuel_df.index:
    brand = fuel_df.loc[i, 'бренд']
    model = fuel_df.loc[i, 'модель']
    kz.loc[((kz['бренд'] == brand) & (kz['модель'] == model)), 'вид_топлива'] = (
        kz.query('бренд == @brand and модель == @model')['вид_топлива'].mode()[0])
```

```python
len(kz.groupby(['бренд', 'модель'], as_index=False)
           .agg({'вид_топлива': 'nunique'})
           .query('вид_топлива > 1'))
```

```python
kz['вид_топлива'].unique()
```

Таким образом исправили значение топлива "0" и кроме категорий остались только пропуски.

**Объем двигателя**

```python
kz['объем_двиг_л'].unique()
```

Укажем корректное значение разделитя и удалим ненужные буквенные обозначение, там где есть корректное значение объема двигателя.

```python
kz['объем_двиг_л'].unique()
```

```python
kz['объем_двиг_л'] = pd.to_numeric(kz['объем_двиг_л'], errors='coerce')
```

```python
kz['объем_двиг_л'].unique()
```

Посмотрим есть ли модели авто, для которых указано более одного значения объема двигателя, а также его минимальное, максимальное и медианное значения, чтобы определить ошибочные значения и оценить возможность исправить их.

```python
kz.query('бренд == "Volkswagen" and модель == "Polo"')['объем_двиг_л'].unique()
```

```python
kz.loc[((kz['бренд'] == 'Volkswagen') & (kz['модель'] == 'Polo')), 'объем_двиг_л'] = (
    kz.query('бренд == "Volkswagen" and модель == "Polo"')['объем_двиг_л'].mode()[0])
```

**Тип трансмиссии**

В зависимости от коробки передач трансмиссия бывает: 
- механическая;
- автоматическая;
- роботизированная;
- вариативная (бесступенчатая), т.е. с вариатором.

Разделим тип коробки передач по этим 4 категориям. Для начала посмотрим какие значение есть в столбце и приведем их к одному формату, чтобы убрать неявные дубликаты.

```python
kz['коробка_передач'].unique()
```

```python
kz['коробка_передач'] = kz['коробка_передач'].str.upper().str.replace('/', '')
```

```python
kz['коробка_передач'].unique()
```

Есть несколько значений, которые нельзя отнести ни к одной категории - соберем их в отдельный список и посмотрим к каким брендам и моделям они относятся.

```python
nn = ['8', '0', '#НД', 'РЕДУКТОР', 'TDI']
kz.query('коробка_передач in @nn').groupby(['бренд', 'модель']).agg({'год_выпуска': 'unique', 
                                                                     'вид_топлива': 'unique',
                                                                    'объем_двиг_л': 'unique',
                                                                    'коробка_передач': 'unique'})
```

Согласно открытым данным (интернет), Jac T6 может иметь только механическую коробку передач. Указанные модели Ягуара и Порше - это АКПП, а вот Рено и Тойота могут быть как с механикой, так и с автоматом, поэтому будем их заменять на nan.

```python
def transmission(row):
    '''Категоризация по типу коробки передач'''
    
    AT = ['АКП', 'AT', 'АТ', 'АT', 'АКПП', '6A', '6А', '8A', 'TIPTRONIC', 'STEPTRONIC', 'G-TRONIC', 'PDK', 'РЕДУКТОР', '8']
    MT = ['МТ', 'MT', 'МT', 'МКП', 'МКПП', 'МЕХ.', '5M', '5М', 'TDI']
    VAR = ['ВАРИАТОР', 'CVT']
    robot = ['DSG', 'S-TRONIC', 'DCT', 'POWERSHIFT S6']
    
    if any(word in row['коробка_передач'] for word in VAR):
        return 'вариативная'
    if any(word in row['коробка_передач'] for word in AT):
        return 'АКПП'
    if any(word in row['коробка_передач'] for word in MT):
        return 'МКПП'
    if any(word in row['коробка_передач'] for word in robot):
        return 'роботизированная'
    else:
        pass
```

```python
kz['коробка_передач'] = kz['коробка_передач'].astype('str')
kz['коробка_передач'] = kz.apply(transmission, axis=1)
```

```python
kz['коробка_передач'].unique()
```

**Тип привода**

Выделим следующие категории в столбце с типом привода:
- RWD – задний привод;
- FWD – передний привод;
- 4WD – полный привод;

Для начала приведем все значения в столбце к одному формату (верхний регистр, т.к. есть аббривиатуры), уберем лишние пробелы и сделаем формат данных строковый для корректной обработки.

```python
kz['тип_привода'] = (kz['тип_привода'].str.upper()
                     .str.replace(' ', '')
                     .str.replace('Х', 'X'))
kz['тип_привода'] = kz['тип_привода'].astype('str')
```

Некорректные значения будем заменять на пропуски.

```python
def privod(row):
    '''Категоризация по типу привода'''
    
    RWD = ['ЗАДНИЙ', 'RWD']
    FWD = ['ПЕРЕДНИЙ', 'FWD', 'FF']
    AWD = ['QUATTRO', 'ПОЛНЫЙ', '4WD', 'AWD', '4X4', '4MOTION']
    TWO_WHEEL = ['4X2.2', '2WD', '4X2'] 
    
    if any(word in row['тип_привода'] for word in RWD):
        return 'RWD'
    if any(word in row['тип_привода'] for word in FWD):
        return 'FWD'
    if any(word in row['тип_привода'] for word in AWD):
        return '4WD'
    if any(word in row['тип_привода'] for word in TWO_WHEEL):
        return '2WD'
    else:
        pass
```

#### Динамика продаж в целом по рынку: в денежном и натуральном выражении (помесячно)

Сумма продаж авто за первые 9 месяцев 2019 года имела следующую динамику: после небольшого уменьшения на второй месяц, сумма продаж уверенно росла с февраля по май, далее незначительное уменьшение в июне и резкий пик продаж в июле.

```python
kz.groupby('месяц').agg({'количество': 'sum'}).plot(figsize=(8,4), legend=False)
plt.title('Динамика количества проданных авто по месяцам')
plt.ylabel('Количество')
plt.xlabel('Месяц')
plt.show();
```

Динамика количества проданных авто по месяцам практически повторяет динамику суммы продаж: уверенный рост с февраля по мая, минимальные изменения в июне и резкий пик продаж в июле.

#### Динамика продаж по категориям (помесячно): тип топлива, класс, сегмент

```python
cat = ['вид_топлива', 'сегментация_2013']

for col in cat:
    
    kz.pivot_table(columns=col, index='месяц', values='продажа_usd', aggfunc='sum').plot(figsize=(8,4))
    plt.legend(title=col, bbox_to_anchor=(1,1), loc='upper left')
    plt.title('Динамика суммы продаж помесячно по '+col)
    plt.xlabel('Месяц')
    plt.ylabel('Сумма продаж, USD')
    plt.show();
```

Подавляющее число проданных авто работают на бензине, поэтому динамика суммы продаж на бензине повторяет общий график динамики суммы продаж помесячно. 

Наибольшую прибыль приносят продажи внедорожников, а также легковых и коммерческих авто. При чем июльский пик продаж более выражен для коммерческих авто. Минивэны и пикамы - наименее популярные сегменты.

```python
top_class = (kz.groupby('класс_2013', as_index=False)
             .agg({'продажа_usd': 'sum'})
             .sort_values(by='продажа_usd', ascending=False)['класс_2013'].head(10))

df_top_class = kz.query('класс_2013 in @top_class')
df_top_class['класс_2013'] = df_top_class['класс_2013'].astype('str')

df_top_class.pivot_table(columns='класс_2013', index='месяц', values='продажа_usd', aggfunc='sum').plot(figsize=(8,4))
plt.legend(title='класс', bbox_to_anchor=(1,1), loc='upper left')
plt.title('Динамика суммы продаж помесячно для топ-10 классов')
plt.xlabel('Месяц')
plt.ylabel('Сумма продаж, USD')
plt.show();
```

Среди топ-10 классов авто по сумме продаж за первые 9 месяцев наибольшую прибыль приносят продажи компактных и среднеразмерных внедорожников, а также авто класса D (большой семейный автомобиль). В январе, феврале и с июня по сентябрь лидерство по продажам удерживали авто класса компактные SUV. В марте лидировали авто D класса, а в апреле и мае - среднеразмерные SUV. Также можно отметить резкий рост спроса на большие автобусы, крупнотонажные грузовики, субкомпактные SUV и авто С класса в июле.

```python
cat = ['вид_топлива', 'сегментация_2013']
for col in cat:
    kz.pivot_table(columns=col, index='месяц', values='количество', aggfunc='sum').plot(figsize=(8,4))
    plt.legend(title=col, bbox_to_anchor=(1,1), loc='upper left')
    plt.title('Динамика количества проданных авто помесячно по '+col)
    plt.xlabel('Месяц')
    plt.ylabel('Количество проданных авто')
    plt.show();
```

По количеству продаваемых авто лидируют внедорожники и легковые авто на бензине. Не смотря на то, что кол-во проданных легковых автомиюелей в некоторые месяцы даже превосходит кол-во проданных внедорожников, наибольшую прибыль все равно приносят внедорожники, вероятно т.к. цена на них выше, чем на легковые.

```python
top_class = (kz.groupby('класс_2013', as_index=False)
             .agg({'количество': 'sum'})
             .sort_values(by='количество', ascending=False)['класс_2013'].head(10))

df_top_class = kz.query('класс_2013 in @top_class')
df_top_class['класс_2013'] = df_top_class['класс_2013'].astype('str')

df_top_class.pivot_table(columns='класс_2013', index='месяц', values='количество', aggfunc='sum').plot(figsize=(8,4))
plt.legend(title='класс', bbox_to_anchor=(1,1), loc='upper left')
plt.title('Динамика количества проданных авто помесячно для топ-10 классов')
plt.xlabel('Месяц')
plt.ylabel('Количество проданных авто')
plt.show();
```

Абсолютный лидер по кол-ву проданных авто - машины класса B. Второе место делят компактные и субкомпактные SUV - последние имеют ярко выраженный пик в июле.
